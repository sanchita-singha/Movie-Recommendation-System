```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
clean_movies = pd.read_csv("C:\\Users\\sanch\\OneDrive\\Desktop\\Movie-Recommendation-System\\data\\clean_movies.csv")
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
clean_movies['genres'] = clean_movies['genres'].apply(convert)
clean_movies.head()
```

```python
clean_movies['keywords'] = clean_movies['keywords'].apply(convert)
clean_movies.head()
```

```python
def convert_cast(text):
    L = []
    counter = 0

    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break

    return L
```

```python
clean_movies['cast'] = clean_movies['cast'].apply(convert_cast)
clean_movies.head()
```

```python
clean_movies['cast'] = clean_movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []

    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])

    return L
```

```python
clean_movies['crew'] = clean_movies['crew'].apply(fetch_director)
```

```python
clean_movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
clean_movies['cast'] = clean_movies['cast'].apply(collapse)
clean_movies['crew'] = clean_movies['crew'].apply(collapse)
clean_movies['genres'] = clean_movies['genres'].apply(collapse)
clean_movies['keywords'] = clean_movies['keywords'].apply(collapse)
```

```python
clean_movies.head()
```

```python
clean_movies['overview'] = clean_movies['overview'].apply(lambda x:x.split())
```

```python
clean_movies['tags'] = clean_movies['overview'] + clean_movies['genres'] + clean_movies['keywords'] + clean_movies['cast'] + clean_movies['crew']
```

```python
clean_movies['tags'] = clean_movies['tags'].apply(lambda x: " ".join(x))
clean_movies.head()
```
